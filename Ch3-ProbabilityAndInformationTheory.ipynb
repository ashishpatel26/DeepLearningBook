### 3.1 Why Probability?

Because we are dealing with uncertain quantities. We use **Probability** to describe some concepts like "degree of believe" or "hapening frequency".

The uncertainty comes from:
- Inherent stochasticity in the system being modeled.
- Incomplete observability.
- Incomplete modeling.


### 3.2 Random Variables
A **random variable** is a variable that can take on different values randomly. Random variables may be discrete or continuous. A discrete random variable is one that has a finite or countably infinite number of states. Note that these states are not necessarily the integers; they can also just be named states that are not considered to have any numerical value. A continuous random variable is associated with a real value.

### 3.3 Probability Distributions
A probability distribution is a description of how likely a random variable or set of random variables is to take on each of its possible states. The way we describe probability distributions depends on whether the variables are discrete or continuous.

- Discrete: Probability Mass Function
- Continuous: Probability Density Function

#### 3.3.1 Discrete Variables and Probability Mass Functions
A probability distribution over discrete variables may be described using a probability mass function (PMF). 

X: All possible states, or we say X is a property
x: one of the state

 ** ∀x∈X, 0 ≤ P(x) ≤ 1 **
 
 If the property is normalized, ∑<sub>x∈X</sub>P(x) = 1

P(x) means the Probability of x event. If x is impossible, P(x<sub>1</sub>) = 0; If x is always true, P(x<sub>1</sub>) = 1.

**Joint Probability** <br>
  To describe the probability that x<sub>1</sub> and x<sub>2</sub> both happened simotaneously, we call it joint porbability, P(x<sub>1</sub>,x<sub>2</sub>).

**Uniform distribution**
  If we got k elements, every elements have same probability. Because ∑P(x) = 1, ==> P(x) = 1/k.

#### 3.3.2 Continuous Variables and Probability Density Functions
When working with continuous random variables, we describe probability distri- butions using a probability density function (PDF).

X: All possible states
x: one of the state

 ** ∀x∈X, P(x) ≥ 0 and ∫P(x)dx = 1 **




### 3.4 Marginal Probability

The probability distribution over the subset is known as the marginal probability distribution. For example: P(a=1,b=1) means the probability that *"a=1 and b=1"*.

### 3.5 Conditional Probability

The probability when given some events is called **Conditional Probability**. For example, The probability of train delayed when rainy day is described as P(Delay | Rainy Day).

** Conditional Probability: P(x | Given Condition) **

Here is one important rule:
** P(x,y) = P(x|y)∙P(y) **

The probability that train delayed and rainy is the probability of rainy day times the probability of train delayed at rainy day.


### 3.6 The Chain Rule of Conditional Probabilities

P(x<sup>(1)</sup>,x<sup>(2)</sup>,...,x<sup>(n)</sup>) <br> 
  = P(x<sup>(n)</sup>|x<sup>(1)</sup>,...,x<sup>(n-1)</sup>)∙P(x<sup>(1)</sup>,...,x<sup>(n-1)</sup>) <br>
  = P(x<sup>(n)</sup>|x<sup>(1)</sup>,...,x<sup>(n-1)</sup>)∙P((x<sup>(n-1)</sup>|x<sup>(1)</sup>,...,x<sup>(n-2)</sup>)∙P(x<sup>(1)</sup>,...,x<sup>(n-2)</sup>) <br>
  = ...<br>
  = P(x<sup>(1)</sup>) ∏<sub>i=2</sub><sup>n</sup>P(x<sup>(i)</sup>|x<sup>(1)</sup>,...,x<sup>(i-1)</sup>)
  
∏: Multiply

For example: <br>
P(x,y,z) = P(x|y,z)∙P(y,z)= P(x|y,z)∙P(y|z)∙P(z) <br>
P(x,y,z) = P(z|x,y)∙P(x,y)= P(z|x,y)∙P(y|x)∙P(x) <br>


### 3.7 Independence and Conditional Independence

We say 2 variables x,y are **independent** : P(x,y) = P(x)∙P(y)

We say 2 variables x,y are independent given z, that is **Conditional Independent** :<br>
P(x,y,z) = P(x|z)∙P(y|z)


### 3.8 Expectation, Variance and Covariance

Expectation (Expected Value) E =∑<sub>x</sub>P(x)𝑓(x)  or ∫p(x)𝑓(x)dx <br>

Expectation is linear: E[a∙𝑓(x) + b∙𝑔(x)] = a∙E[𝑓(x)] + b∙E[𝑔(x)]

**Variance** is Measure of how much the values of a function of a random variable x vary as we sample diﬀerent values of x from its probability distribution.<br>

  Variance: Var(𝑓(x)) = E[(𝑓(x) - E[𝑓(x)])<sup>2</sup>]

  Standard Deviation: Std(𝑓(x)) = Var(𝑓(x))<sup>1/2</sup>

**Covariance** gives some sense of how much two values are linearly related to each other.

  Cov(𝑓(x),𝑔(x)) = E[(𝑓(x) - E[𝑓(x)])∙(𝑔(x) - E[𝑔(x)])]

  Covariance 絕對值很大代表的意義是：當一個變數偏離平均時，另一個變數也傾向偏離平均。
  Covariance 負很大時，當一個變數偏離平均時，另一個變數傾向往另外一邊偏離平均。
  
**Correlation** normalize the contribution of each variable in order to measure only how much the variables are related, rather than also being aﬀected by the scale of the separate variables. 只關注兩個變數的相關程度，而排除了這兩個變數的尺度大小差異。


**Covariance and Dependency** <br>
  Does Cov()=0 means Independency? (Nope, because...)
    - For two variables to have zero covariance, there must be no linear dependence between them. But, maybe they are nonlinear dependent.<br>
    - Independenceis a stronger requirement than zero covariance, because independence also excludesnonlinear relationships.<br>
    



### 3.9 Common Probability Distributions

#### 3.9.1 Bernoulli Distribution
  A distribution over a single binary random variable. φ∈[0,1]

#### 3.9.2 Multinoulli Distribution
  The multinoulli or categorical distribution is a distribution over a single discrete variable with k diﬀerent states, where k is ﬁnite.
  
  **p∈[0,1]<sup>k-1</sup>**, where p<sup>i</sup> gives the probability of the i-th state.
  
  The multinoulli distribution is a special case of the multi-nomial distribution. A multi-nomial distribution is the distribution over vectors in{0,..., n}<sup>k</sup> representing how many times each of the k categories is visited when n samples are drawn from a multinoulli distribution.
  
#### 3.9.3 Gaussian Distribution (Normal Distribution)


\begin{align}
  N(x;\mu,\sigma^2) &= \dfrac{1}{\sigma \sqrt{2\pi}} e^{-\dfrac{(x-\mu)^2}{2 \sigma^2}} 
\end{align}

The two parameters µ ∈ R and σ ∈(0, ∞) control the normal distribution.    
    
#### 3.9.4 Exponential and Laplace Distributions

Exponential distribution is popular for deep learning algorithm.
\begin{align*}
  p(x;\lambda) &= \lambda e^{-\lambda x}, when x > 0 \\
  p(x;\lambda) &= 0, when x ≤ 0
\end{align*}


_Laplace  distribution_

$$ Laplace(x; \mu, \gamma) = \dfrac{1}{2\gamma}e^{-\dfrac{| x-\mu |}{\gamma}}
$$

_Poisson distribution_
$$ P(X=k) = \dfrac{e^{-\lambda}\lambda^k}{k!}
$$

where
  - lambda is the average number of events per interval
  - k takes values 0, 1, 2, …

#### 3.9.5 The Dirac Distribution and Empirical Distribution

The Dirac delta function is deﬁned such that it is zero-valued everywhere except 0, yet integrates to 1.

\begin{align*}
  \text{Dirac delta function: } \delta(x)
  p(x) = \delta(x-\mu)
\end{align*}

A common use of the Dirac delta distribution is as a component of an empirical distribution

$$ \hat{p}(x)=\dfrac{1}{m} \sum_{i=1}^m \delta (x - x^{(i)}) 
$$

an **empirical distribution** can be conceptualized as a multinoulli distribution, with a probability associated to each possible input value that is simply equal to the empirical frequency of that value in the training set.

#### 3.9.6 Mixtures of Distributions

Misture distribution is to combine different distributions,

$$ P(x) = \sum_{i} P(x|c=i)P(c=1)
$$

A very powerful and common type of mixture model is the **Gaussian mixture model**, in which the components p(x | c=i) are Gaussians.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

mu = 5
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(0, 10, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma),color='r')
plt.text(6,0.3,'avg=5,var=1')

plt.plot(x,mlab.normpdf(x, mu, (2*sigma)),color='b')
plt.text(7,0.13,'avg=5,var=2')

plt.title('Gaussian Distribution')
plt.xlabel('f(x)')
plt.ylabel('Probability')
plt.show()
```

### 3.10 Useful Properties of Common Functions

$$ 
\text{Logistic Sigmoid Function } \sigma (x) = \frac{1}{1+e^{-x}}
$$
  
$$
\text{Softplus Function } \zeta(x) = log(1+e^{x}) 
$$


```python
import numpy as np
import matplotlib.pyplot as plt

def dirac(x):
    if ((x < -0.01)|(x > 0.01)):
        return 0.0
    else:
        return 1.0

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softplus(x):
    return np.log(1 + np.exp(x))

if __name__ == "__main__":
    x = []
    y = []
    z = []
    di = []
    for i in np.arange(-5.0,5.0,0.1):
        x.append(i)
        y.append(sigmoid(i))
        z.append(softplus(i))
        di.append(dirac(i))

    fig = plt.figure()
    fig.add_subplot(3,1,1)
    plt.title('Dirac Function')
    plt.grid(True)
    plt.plot(x,di)

    fig.add_subplot(3,1,2)
    plt.title('Sigmoid Function')
    plt.grid(True)
    plt.plot(x,y)

    fig.add_subplot(3,1,3)
    plt.title('Softplus Function')
    plt.grid(True)
    plt.plot(x,z)
    fig.tight_layout()
    plt.show()
```

### 3.13 Information Theory

Information theory is to quantify **how much** information is present in a signal.

- Likely events should have low information content. Guaranteed events have no information.
- Less likely events should have higher information content.
- Independent events should have additive information. For example, ﬁnding out that a tossed coin has come up as heads twice should convey twice as much information as ﬁnding out that a tossed coin has come up as heads once.

Define **self-information** of an event x = 𝑥 to be 

$$ I(x) = −log P(x)
$$

The unit I(x) is called "nats". One nat is the amount of information gained by observing an event of probability e<sup>-1</sup>. Some textbooks called it "bits" or "shannons".  

Self-information deals only with a single outcome. We can quantify the amountof uncertainty in an entire probability distribution using the **Shannon entropy**:

$$ H(P) = H(x) = E_{x∼P}[I(x)] = −E_{x∼P}[log P(x)]
$$

The Shannon entropy of a distribution is the expected amount of information in an event drawn from that distribution. It givesa lower bound on the number of bits (if the logarithm is base 2, otherwise the units are diﬀerent) needed on average to encode symbols drawn from a distribution P. When x is continuous, the Shannon entropy is known as the **diﬀerential entropy**.  


If we have two separate probability distributions P(x) and Q(x) over the same random variable x, we can measure how diﬀerent these two distributions are using the **Kullback-Leibler (KL) divergence**:  

$$ D_{KL}(P||Q) = E_{x∼P}[log\dfrac{P(x)}{Q(x)}] = E_{x∼P}[log P(x) − log Q(x)]
$$



```python
import numpy as np
import matplotlib.pyplot as plt

dx = []
dy = []
for p in np.arange(0.0,1.01,0.01):
    dx.append(p)
    if ((p == 0)|(p == 1)):
        dy.append(0)
    else:
        dy.append((p-1)*np.log(1-p) - p*np.log(p))

fig = plt.figure()
plt.text(0.3,0.55,'Max entropy at p= 0.5')
plt.ylabel('Shannon entropy (nats)')
plt.xlabel('Probability p')
plt.grid(True)
plt.plot(dx,dy)
plt.show()
```

### 3.14 Structured Probabilistic Models


Machine learning algorithms often involve probability distributions over a very large number of random variables. Instead of using a single function to represent a probability distribution, we can split a probability distribution into many factors that we multiply together. We call it a **structured probabilistic model** or **graphical model**.

$$p(a,b,c) = p(a) \cdot p(b|a) \cdot p(c|b)
$$

**Direct Model** use graphs with directed edges, and they represent factorizations into conditional probability distributions.

<img src="DirectModelGraph3.png" alt="Direct Model" style="width: 50%;"/>

$$ p(x) = \prod_{i} p(x | P_{given-parent}(x_i))
$$

$$ p(a,b,c,d,e) = p(a) \cdot p(b) \cdot p(c|a,b) \cdot p(d|a,c) \cdot p(e|c)
$$


**Undirect Model** use graphs with undirected edges, and they represent factorizations into a set of functions. These functions are usually not probability distributions of any kind.  

Define clique C, that all notes connected to each other. (集團內任何兩個頂點都是相鄰的)

Each clique C(i) in an undirected model is associated with a factor φ(i)(C(i)). A factor is a function.

There is no guarantee that sum of these products will be 1. Therefore, we divide it by a normalizing constant Z, deﬁned to be the sum of all φ functions, in order to obtain a normalized probability distribution.

<img src="UndirectModelGraph.png" alt="Direct Model" style="width: 50%;"/>

$$ p(x) = \dfrac{1}{Z} \prod_{i} φ^{(i)}(C^{(i)})
$$

$$ Z = \sum_{i} φ^{(i)}(C^{(i)})
$$

$$ p(a, b, c, d, e) = \dfrac{1}{Z} φ^{(1)}(b, c)φ^{(2)}(a, c, d)φ^{(3)}(c, e)
$$
